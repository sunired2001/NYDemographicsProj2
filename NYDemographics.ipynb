```python
import pandas as pd
from splinter import Browser

import numpy as np
from sodapy import Socrata
from config import app_token
import pymongo
import json
client = Socrata("data.cityofnewyork.us", app_token)
```

```python
#https://catalog.data.gov/dataset/demographic-statistics-by-zip-code-acfc9
NYDemographs= client.get("kku6-nxdu",limit=2000000)
NYDemographsdf=pd.DataFrame.from_records(NYDemographs)
#NYDemographsdf.to_csv("c:/webscraping/NYDemographs.csv")
NYDemographsfinaldf = NYDemographsdf
NYDemographsfinaldf
```

```python
NYDemographsfinaldffilter=pd.DataFrame({})
NYDemographsfinaldffilter['jurisdiction_name']=NYDemographsfinaldf['jurisdiction_name']

NYDemographsfinaldffilter=NYDemographsfinaldffilter.join(NYDemographsfinaldf[[ 'count_participants',"count_male",
                                             "count_gender_unknown","count_gender_total","count_pacific_islander","count_hispanic_latino","count_american_indian","count_asian_non_hispanic","count_white_non_hispanic",
                                             "count_black_non_hispanic","count_other_ethnicity","count_ethnicity_unknown","count_ethnicity_total","count_permanent_resident_alien","count_us_citizen",
                                             "count_other_citizen_status","count_citizen_status_unknown","count_citizen_status_total","count_receives_public_assistance",
                                             "count_nreceives_public_assistance","count_public_assistance_unknown","count_public_assistance_total"]])
NYDemographsfinaldffilter.rename(columns={"jurisdiction_name":"ZipCode"},inplace=True)
NYDemographsfinaldffilter
```

```python
url="https://www.zipcodestogo.com/New%20York/"
zipcodedf=pd.read_html(url,header=0)
zipcodedf1=zipcodedf[1].iloc[1:]
zipcodedf1.columns=["ZipCode","City","County","ZipcodeMap"]
```

```python
zipcodedf1.to_csv("c:/webscraping/ziplist.csv")
zipcodedf1.head()
```

```python
 executable_path = {'executable_path': 'chromedriver.exe'}
 browser = Browser('chrome', **executable_path)
```

```python

browser.visit(url)
viewmap=browser.find_by_css(' a[href*="maps.google.com"] ')
#viewmap=browser.find_by_css('table[class="inner_table"] tbody tr td div[align="center"] a ')
zipurl=[]
for i in viewmap:
    zipurl.append(i["href"])
    
```

```python
zipcodedf1["ZipUrl"]=zipurl
finalZipdata=zipcodedf1.drop(["ZipcodeMap"],axis=1)
finalZipdata
```

```python
NYDemobyCounty=pd.merge(finalZipdata,NYDemographsfinaldffilter,on="ZipCode")
```

```python
NYDemobyCounty.to_excel("NYDemographsByCountt.xlsx")
```
